Quando criamos uma tupla, normalmente atribuímos valores a ela. Isso é chamado de "embalagem" uma tupla:


Embalando uma tupla:

```python
nomes =  ('Jane', 'John', 'jimmy')
```

Mas, em Python, também podemos extrair os valores de volta em variáveis. Isso se chama "desempacotar":

```python
nomes = ('Jane', 'John', 'Jimmy')

(nome1, nome2, nome3) = nomes

print(nome1)
print(nome2)
print(nome3)
```

Se o número de variáveis for menor que o número de valores, você pode adicionar um " * " ao nome da variável e os valores serão atribuídos à variável como uma lista:



```python
nomes = ('Jane', 'Mark', 'John', 'Petter', 'Bill', 'Joseph')

(nome1, nome2, *nome3) = nomes 

print(nome1)
print(nome2)
print(nome3)
```

Se o asterisco for adicionado a outro nome de variável que o não o último, python atribuirá valores à variável até que o número de valores restantes corresponda ao número de variáveis restantes.


**Exemplo**

Adicione uma lista de valores da variável "trópico":

```python
frutas = ('maçã', 'manga', 'mamão', 'cereja', 'morango')

(verde, *tropical, vermelha) = frutas

print(verde)
print(tropical)
print(vermelha)
```
