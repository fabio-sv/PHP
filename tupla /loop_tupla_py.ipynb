# Loop através de uma Tupla

Você pode iterar os itens de uma tupla usando um loop.`for`

**Exemplo**

Iterar através dos itens e imprimir os valores:


```python
tupla = ('Volvo', 'BMW', 'Ford', 'VW')

for listar in tupla:
    print(listar)
```

Você também poder realizar loop através dos itens tupla referindo-se ao seu número de índice.

Use a funções `range()len()` para criar uma iterável adequada:

**Exemplo**

Imprima todos os itens referindo-se ao número do índice:

```python
marcas =  ('Nike', 'Adidas', 'Mizuno', 'Vans', 'Puma')

for listar in range(len(marcas)):
  print(listar)
```

# Usando um loop while

Você pode fazer loop através dos itens da lista usando um loop `while`.

Use a função para determinar o comprimento da tupla, em seguida, comece em 0 e dê a volta por seus itens de tupla, referindo-se aos seus índices `len()`.

**Exemplo**

Imprima todos os itens, usando um loop `while` para analisar todos os números do índice:


```python
tupla = ('Nike', 'Adidas', 'Puma')

i = 0

while i < len(tupla):
  print(tupla[i])
  i = i + 1
```
