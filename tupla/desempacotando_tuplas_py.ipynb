Quando criamos uma tupla, normalmente atribuímos valores a ela. Isso é chamado de "embalagem" uma tupla:


Embalando uma tupla:

```python
nomes =  ('Jane', 'John', 'jimmy')
```

Mas, em Python, também podemos extrair os valores de volta em variáveis. Isso se chama "desempacotar":

```python
nomes = ('Jane', 'John', 'Jimmy')

(nome1, nome2, nome3) = nomes

print(nome1)
print(nome2)
print(nome3)
```
